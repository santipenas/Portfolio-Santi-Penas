## Vector Algebra and Image Manipulation with Python

#### Summary: 
This project explores vector algebra concepts using NumPy and demonstrates image manipulation and visualization. It includes operations such as image inversion, rotation, and solving linear equations, enhanced with exploratory data analysis and statistical insights. And we will se as result some images Transformations with NumPy regarding vector and algebra advance maths.

#### Workflow Process:
1. Define the original image as a matrix of pixel intensities.
2. Apply image inversion using matrix subtraction.
3. Rotate the image using matrix transposition.
4. Generate a random matrix to simulate a linear system.
5. Solve the linear equation to compute X.
6. Reconstruct the heart image using matrix multiplication.
7. Perform EDA to understand the properties of each image.

----

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### We define the heart image using array() method from numpy

```python
heart_img = np.array([
    [255, 0, 0, 255, 0, 0, 255],
    [0, 255/2, 255/2, 0, 255/2, 255/2, 0],
    [0, 255/2, 255/2, 255/2, 255/2, 255/2, 0],
    [0, 255/2, 255/2, 255/2, 255/2, 255/2, 0],
    [255, 0, 255/2, 255/2, 255/2, 0, 255],
    [255, 255, 0, 255/2, 0, 255, 255],
    [255, 255, 255, 0, 255, 255, 255]
])
```

##### This would be our function to display images later

```python
def show_image(image, title):
    """Displays an image with a title."""
    plt.imshow(image, cmap="gray")
    plt.title(title)
    plt.colorbar(label="Pixel Intensity")
    plt.show()
```

##### Visualize the original heart image

```python
show_image(heart_img, "Heart Image")
```

##### Now we can invert the image

```python
inverted_heart_img = 255 - heart_img
show_image(inverted_heart_img, "Inverted Heart Image")
```

##### Also using a transpose operation, we can rotate the image. LetÂ´s have a look:

```python
rotated_heart_img = heart_img.T
show_image(rotated_heart_img, "Rotated Heart Image")
```

##### Generate a random image for a linear algebra operations

```python
np.random.seed(42)  # Ensure reproducibility
random_img = np.random.randint(0, 255, (7, 7))
show_image(random_img, "Random Image")
```

##### Solve for X using linear equations

```python
try:
    X = np.linalg.solve(random_img, heart_img)
    show_image(X, "Solved X Image")
except np.linalg.LinAlgError as e:
    print("Error solving linear equations:", e)
```

##### Finally, we reconstruct the original heart image

```python
try:
    solved_heart_img = random_img @ X
    show_image(solved_heart_img, "Reconstructed Heart Image")
except Exception as e:
    print("Error reconstructing the image:", e)
```

##### To conclude, let go deeper into a basic Exploratory Data Analysis

```python
def image_statistics(image, title):
    """Calculate and display statistics for an image."""
    print(f"--- {title} Statistics ---")
    print(f"Min Intensity: {np.min(image)}")
    print(f"Max Intensity: {np.max(image)}")
    print(f"Mean Intensity: {np.mean(image):.2f}")
    print(f"Median Intensity: {np.median(image):.2f}")
    print(f"Standard Deviation: {np.std(image):.2f}")
```

```python
image_statistics(heart_img, "Original Heart Image")
image_statistics(inverted_heart_img, "Inverted Heart Image")
image_statistics(rotated_heart_img, "Rotated Heart Image")
image_statistics(random_img, "Random Image")
```
