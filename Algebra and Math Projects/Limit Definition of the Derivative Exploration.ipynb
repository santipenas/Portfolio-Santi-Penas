# Numerical Differentiation and Analysis Using Python

## Introduction

This notebook explores numerical differentiation using the limit definition of derivatives. It calculates derivatives of functions at given points, analyzes their behavior, and visualizes the convergence of approximated derivatives to the true derivative by varying step sizes (`h`). 

### Objectives:
1. Implement numerical differentiation using Python.
2. Visualize the relationship between step size (`h`) and derivative accuracy.
3. Explore the behavior of functions and their derivatives.
4. Provide insights and conclusions based on the analysis.



## Workflow Process

1. **Define the Target Functions**: Functions for which derivatives will be computed are defined.
2. **Implement Numerical Differentiation**: The `limit_derivative` function calculates the derivative using the limit definition.
3. **Visualization**: The convergence of approximated derivatives is visualized for various step sizes.
4. **Insights and Conclusions**: Analyze and interpret results, focusing on convergence trends and derivative accuracy.

---


```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from math import sin, cos, log, pi
```

#### Define Functions and Numerical Differentiation

```python
# Numerical Differentiation Function
def limit_derivative(f, x, h):
        return (f(x + h) - f(x)) / h

# Target Functions
def f1(x):
    return sin(x)

def f2(x):
    return pow(x, 4)

def f3(x):
    return pow(x, 2) * log(x)
```

#### Step-by-Step Derivative Calculation

```python
# Compute derivatives for f3 at x=1 with varying h values
h_values = [2, 0.1, 0.00001]
for h in h_values:
    derivative = limit_derivative(f3, 1, h)
    print(f"Derivative of f3 at x=1 with h={h}: {derivative}")
```

#### Visualizing Derivative Convergence

```python
# Plotting Convergence
def plot_approx_deriv(f, function_name):
    x_vals = np.linspace(1, 10, 200)
    h_vals = [10, 1, 0.25, 0.01]
    
    plt.figure(figsize=(10, 6))
    for h in h_vals:
        derivative_values = [limit_derivative(f, x, h) for x in x_vals]
        plt.plot(x_vals, derivative_values, linestyle='--', label=f"h={h}")
    
    plt.title(f"Convergence of Approximate Derivatives for {function_name}")
    plt.xlabel("x")
    plt.ylabel("Derivative Value")
    plt.legend()
    plt.grid()
    plt.show()

# Plot for f1
plot_approx_deriv(f1, "sin(x)")

# Plot for f2
plot_approx_deriv(f2, "x^4")
```

#### Enhanced Visualizations for True Derivatives

```python
# True Derivatives
x_vals = np.linspace(1, 10, 200)

# True derivative for sin(x)
y_vals_sin = [cos(val) for val in x_vals]
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals_sin, label="True Derivative of sin(x)", linewidth=2, color='red')
plt.title("True Derivative of sin(x)")
plt.xlabel("x")
plt.ylabel("Derivative Value")
plt.legend()
plt.grid()
plt.show()

# True derivative for x^4
y_vals_x4 = [4 * pow(val, 3) for val in x_vals]
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals_x4, label="True Derivative of x^4", linewidth=2, color='blue')
plt.title("True Derivative of x^4")
plt.xlabel("x")
plt.ylabel("Derivative Value")
plt.legend()
plt.grid()
plt.show()
```

## Conclusion

### Key Insights:

1. **Accuracy of Approximation**:  
   - As the step size (`h`) decreases, the numerical derivative converges to the true derivative.  
   - This demonstrates the reliability of the limit definition for approximating derivatives.

2. **Visualization Insights**:  
   - For `sin(x)`, the convergence of numerical derivatives is smooth, and the approximated values align closely with the true derivative as `h` decreases.  
   - For `x^4`, the derivative values also converge, but the convergence rate varies due to the steepness of the function.

3. **Trade-offs Between Step Size and Numerical Errors**:  
   - While reducing `h` improves the accuracy of the approximation, extremely small values of `h` can introduce computational errors due to floating-point precision limits.

### Overall Summary:
This analysis highlights the interplay between mathematical theory and computational implementation. It emphasizes the importance of choosing an appropriate step size to balance accuracy and precision while applying numerical differentiation methods. By visualizing convergence trends, the study provides practical insights into the behavior of derivatives and their approximations.
