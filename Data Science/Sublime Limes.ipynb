```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

### Data Preparation

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]
```

Convert this Data to pandas DataFrame for better analysis. We need to create a new one, we will store this in a variable called *data*.


```python
data = pd.DataFrame({
    'Month': months,
    'Visits': visits_per_month,
    'Key Limes': key_limes_per_month,
    'Persian Limes': persian_limes_per_month,
    'Blood Limes': blood_limes_per_month
})
```

Check the new created dataset

```python
display(data)
```

See a brief summary statistics

```python
print(data.describe())
```

Calculate total lime sales per month


```python
total_limes_sales = data['Key Limes'] + data['Persian Limes'] + data['Blood Limes']
```

```python
print("The total sales per month is:\n")
print(total_limes_sales)
```

Now that we have new results for the total sales, we can add a new column to our dataframe to keep this new info.

```python
data["Total Sales per Month"] = total_limes_sales
```

```python
display(data)
```

Rolling Average for Visits


```python
data['Visits (Rolling Avg)'] = data['Visits'].rolling(window=3).mean()
```

### Data Visualization Lines

1. Plot Visit per Month

```python
plt.plot(data['Month'], data['Visits'], marker='o', linestyle='--', color='blue', label='Visits')
plt.title('Visits per Month', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Visits', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
```

We see that in June we have most of the visitors during the year.

2. Plot Lime Sales


```python
plt.plot(data['Month'], data['Key Limes'], marker='o', linestyle='-', color='green', label='Key Limes')
plt.plot(data['Month'], data['Persian Limes'], marker='o', linestyle='-', color='orange', label='Persian Limes')
plt.plot(data['Month'], data['Blood Limes'], marker='o', linestyle='-', color='red', label='Blood Limes')
plt.title('Lime Sales per Month', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Limes Sold', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()
```

Go over the statistics again after this plots

```python
data.describe()
```

#### Some Insights so far:
1. The website visits peaked in June, likely indicating a seasonal trend.
2. Lime sales are relatively consistent, but Key Limes are the most sold variety overall.
3. Persian Limes have the lowest sales volume, possibly indicating lower popularity or limited supply.

3. Heatmap for Correlation

```python
correlation_data = data.drop(columns=['Month']).corr()
plt.figure(figsize=(16, 16))
plt.subplot(3, 2, 4)
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=14)
```

4. Lastly, we will see a stacked Bar Chart for Lime Sales Composition to look visualy the importance of each kinf of lime in the total volume of sales


```python
plt.bar(data['Month'], data['Key Limes'], label='Key Limes', color='lime', alpha=0.7)
plt.bar(data['Month'], data['Persian Limes'], bottom=data['Key Limes'], label='Persian Limes', color='orange', alpha=0.7)
bottom_bar = np.array(data['Key Limes']) + np.array(data['Persian Limes'])
plt.bar(data['Month'], data['Blood Limes'], bottom=bottom_bar, label='Blood Limes', color='red', alpha=0.7)
plt.title('Lime Sales Composition per Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
```

### Conclusions:
1. Website visits peak in June and gradually decline towards December, highlighting a potential seasonal trend. 
2. Rolling average confirms an upward trend in visits between January and June.
3. Lime sales exhibit consistent performance across months, with Key Limes being the most popular variety.
4. Blood Limes and Persian Limes have stable but lower sales compared to Key Limes.
5. Stacked bar chart reveals that Key Limes dominate sales every month, indicating a strong customer preference
6. Efforts to promote Blood and Persian Limes could help balance inventory and revenue.
7. The correlation heatmap shows:
   - Strong positive correlation between total lime sales and Key Lime sales (likely due to their dominance).
   - No significant correlation between visits and lime sales, suggesting lime sales aren't directly impacted by website traffic.


### BI Recommendations:
- Focus marketing campaigns on Key Limes while finding ways to boost interest in Persian and Blood Limes.
- Capitalize on peak website traffic in June with special offers or promotions.
- Further research is needed to understand the factors driving website visits and their potential impact on lime sales.
